```python
#import matplotlib library
import matplotlib.pyplot as plt
```

```python
#import numpy to get data for our plots
import numpy as np
```

```python
x= np.linspace(0,10,100) #evenly spaced 100 values between 0 and 10
y= np.sin(x) #function gives us value for array
z= np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

Plotting the data

```python
#Sine Wave
plt.plot(x,y)
plt.show() #got sine wave
```

```python
#cosine wave
plt.plot(x,z)
plt.show()
```

```python
#adding title x-axis and y-axis
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine Wave')
plt.show()
```

```python
#plot parabola
x=np.linspace(-10, 10, 20)
y= x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x, y , 'r+')
plt.show()
```

```python
plt.plot(x,y , 'g.')
plt.show()
```

```python
#multiple lines in a single graph
x= np.linspace(-5, 5,50)
plt.plot(x, np.sin(x), 'g-')
plt.plot(x, np.cos(x), 'y--')
plt.show()
```

Barplot

```python
fig= plt.figure()
ax = fig.add_axes([0,0,1,1]) # ax----axis
language = ['English', 'French', 'German', 'Latina', 'Spanish']
people=[100,50,150,40,70]
ax.bar(language, people)
plt.xlabel('LANGUAGE')
plt.ylabel('NUMBER OF PEOPLE')
plt.show()
```

Piechart

```python
fig1= plt.figure()
ax = fig1.add_axes([0,0,1,1])
language = ['English', 'French', 'German', 'Latina', 'Spanish']
people=[100,50,150,40,70]
ax.pie( people, labels=language, autopct= '%1.1f%%')#autopct gives how many floating/decimal values we need while chart is plotted
plt.show()
```
